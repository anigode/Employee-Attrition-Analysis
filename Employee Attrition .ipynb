```python
# Importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Inspect unique values in 'Attrition' and 'OverTime'
print(data['Attrition'].unique())
print(data['OverTime'].unique())
```

```python
# Check for missing values in the target variable
if data['Attrition'].isnull().sum() > 0:
    raise ValueError("Target variable 'Attrition' contains missing values.")
```

```python
# Convert 'Attrition' and 'OverTime' columns to numerical
data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})
data['OverTime'] = data['OverTime'].map({'Yes': 1, 'No': 0})
```

```python
# Handle zero value issues
problematic_columns = ['Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 
                       'MonthlyIncome', 'MonthlyRate', 'PercentSalaryHike', 
                       'TotalWorkingYears', 'YearsAtCompany', 
                       'YearsInCurrentRole', 'YearsSinceLastPromotion', 
                       'YearsWithCurrManager']
```

```python
for column in problematic_columns:
    median_value = data[column].replace(0, np.nan).median()  # Replace 0 with NaN to calculate median correctly
    data[column].replace(0, median_value, inplace=True)
```

```python
# Drop duplicates
data.drop_duplicates(inplace=True)
```

```python
# Define categorical and numerical columns
categorical_columns = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18']
numerical_columns = [col for col in data.columns if col not in categorical_columns + ['Attrition']]
```

```python
# Preprocessing and model pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_columns),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)
    ])

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])
```

```python
# Split features and target variable
X = data.drop('Attrition', axis=1)
y = data['Attrition']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Check the data types of X_train to diagnose issues
print("Data types in X_train:")
print(X_train.dtypes)
```

```python
# Save the cleaned dataset
data.to_csv('Employee_attrition_dataset.csv', index=False)
```

```python
# Calculate attrition rate
attrition_rate = data['Attrition'].mean() * 100
print(f'Attrition Rate: {attrition_rate:.2f}%')
```

```python
# Fit the model using the pipeline
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
```

```python
# Evaluation
print(classification_report(y_test, y_pred))
print(f'Accuracy: {accuracy_score(y_test, y_pred):.2f}')

# Check shapes
print(f'y_test shape: {y_test.shape}, y_pred shape: {y_pred.shape}')
print(f'Sample of y_test: {y_test.head()}')
print(f'Sample of y_pred: {y_pred[:5]}')
```
